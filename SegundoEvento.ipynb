```julia
using SACFiles
```

```julia
S2Z = readsac("Sismo2Z.SAC")
```

```julia
using PyPlot
```

```julia
typeof(S2Z)
```

```julia
S2Z.hdr.delta
```

```julia
D2Z = S2Z.data
```

```julia
plot(D2Z)
```

```julia
ts21 = (collect(0:1:length(D2Z)-1))*S2Z.hdr.delta
```

```julia
plot(ts21,D2Z)
```

```julia
length(D2Z)
```

```julia
ondaP = D2Z[500:1500]
```

```julia
plot(ondaP)
```

```julia
Ftra1 = fft(ondaP)
```

```julia
plot(abs(Ftra1))
```

```julia
plot(abs(Ftra1[1:500]))
```

```julia
ondaPZ = Ftra1[1:500]
```

```julia
FxondaP = collect(0:500-1)*(100/length(Ftra1))
```

```julia
plot((FxondaP),abs(log10(ondaPZ)))
ylabel("Amplitud(Log)")
xlabel("Frecuencia (Hz)")
```
