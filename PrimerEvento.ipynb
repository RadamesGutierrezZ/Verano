```julia
using SACFiles
```

```julia
S1Z = readsac("Sismo1Z.SAC")
```

```julia
using PyPlot
```

```julia
typeof(S1Z)
```

```julia
S1Z.hdr.delta
```

```julia
D1Z = S1Z.data
```

```julia
plot(D1Z)
```

```julia
ts11 = (collect(0:1:length(D1Z)-1))*S1Z.hdr.delta
```

```julia
plot(ts11,D1Z)
```

```julia
length(D1Z)
```

```julia
ondaP = D1Z[2100:2600]
```

```julia
plot(ondaP)
```

```julia
Ftra1 = fft(ondaP)
```

```julia
plot(abs(Ftra1))
```

```julia
plot(abs(Ftra1[1:250]))
```

```julia
ondaPZ = Ftra1[1:250]
```

```julia
FxondaP = collect(0:250-1)*(100/length(Ftra1))
```

```julia
plot(((FxondaP)),abs(log10(ondaPZ)))
ylabel("Amplitud(Log)")
xlabel("Frecuencia (Hz)")
```
